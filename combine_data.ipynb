```python
#import dependencies
import pandas as pd 
```

```python
# Read data from file 'filename.csv' 
led = pd.read_csv("Resources/Data/life_expectancy.csv") 
alc_con = pd.read_csv("Resources/Data/world_alcohol_consumption.csv")
happiness = pd.read_csv("Resources/Data/world_happiness_2020.csv")
# Preview the first 5 lines of the loaded data 
print(led.columns)
print(alc_con.columns)
print(happiness.columns)
```

```python
# Combine the data into a single dataset
led = led.loc[led['Year']==2015]
led['Country'].value_counts()
```

```python
alc_con=alc_con.rename(columns={'country':'Country'})
happiness = happiness.rename(columns={'Country name': 'Country'})
happiness.head()
```

```python
led = led.drop(columns=['Year'])
led.head()
```

```python
alc_con
```

```python
happiness
happiness_clean = happiness[["Country", "Regional indicator", "Ladder score", "Logged GDP per capita", "Social support", "Healthy life expectancy", "Freedom to make life choices", "Generosity", "Perceptions of corruption", "Ladder score in Dystopia"]]
happiness_clean
```

```python
combined_df = pd.merge(alc_con,led, on="Country",how='outer')
combined_df
```

```python
combined_data=pd.merge(combined_df,happiness_clean, on="Country",how='outer')
combined_data
```

```python
combined_data.to_csv('Resources/Data/combined_data.csv')
```

```python
print(combined_data.columns)
```

```python
happiness_data=combined_data[['Country', 'beer_servings', 'spirit_servings', 'wine_servings',
       'total_litres_of_pure_alcohol', 'Lifeexpectancy',
       'AdultMortality', 'BMI','GDP', 'Population','Incomecompositionofresources',
       'Schooling', 'Regional indicator', 'Ladder score',
       'Logged GDP per capita', 'Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption', 'Ladder score in Dystopia']]
```
