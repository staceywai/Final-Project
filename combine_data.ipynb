```python
#import dependencies
import pandas as pd 
```

```python
# Read data from file 'filename.csv' 
led = pd.read_csv("Resources/Data/life_expectancy.csv") 
alc_con = pd.read_csv("Resources/Data/world_alcohol_consumption.csv")
happiness = pd.read_csv("Resources/Data/world_happiness_2020.csv")
# Preview the first 5 lines of the loaded data 
print(led.columns)
print(alc_con.columns)
print(happiness.columns)
```

```python
# Combine the data into a single dataset
led = led.loc[led['Year']==2015]
led['Country'].value_counts()
```

```python
alc_con=alc_con.rename(columns={'country':'Country'})
happiness = happiness.rename(columns={'Country name': 'Country'})
happiness.head()
```

```python
led = led.drop(columns=['Year'])
led.head()
```

```python
alc_con
```

```python
happiness
happiness_clean = happiness[["Country", "Regional indicator", "Ladder score", "Logged GDP per capita", "Social support", "Healthy life expectancy", "Freedom to make life choices", "Generosity", "Perceptions of corruption", "Ladder score in Dystopia"]]
happiness_clean
```

```python
combined_df = pd.merge(happiness,alc_con, on="Country",how='left')
combined_df
```

```python
combined_data = pd.merge(combined_df,led, on="Country",how='left')
combined_data
```

```python
combined_data=combined_data.drop(columns=['Totalexpenditure','Alcohol','Standard error of ladder score',
       'upperwhisker', 'lowerwhisker',  'Explained by: Log GDP per capita', 'Explained by: Social support',
       'Explained by: Healthy life expectancy',
       'Explained by: Freedom to make life choices',
       'Explained by: Generosity', 'Explained by: Perceptions of corruption',
       'Dystopia + residual','Regional indicator', 'Status'])
combined_data
```

```python
column_list = combined_data.columns
for column in column_list:
    blank = combined_data[f'{column}'].isnull().sum()
    print(column, ": ", blank)
```

```python
column_list = combined_data.columns
for column in column_list:
    if column != 'Country':
        avg = combined_data[f'{column}'].mean()
        #print(column, ":", avg)
        combined_data[f'{column}'] = combined_data[f'{column}'].fillna(avg)
combined_data
```

```python
column_list = combined_data.columns
for column in column_list:
    blank = combined_data[f'{column}'].isnull().sum()
    print(column, ": ", blank)
```

```python
combined_data=combined_data.rename(columns={
'beer_servings':'Beer_Servings', 'spirit_servings':'Spirit_Servings', 'wine_servings':'Wine_Servings',
       'total_litres_of_pure_alcohol':'Total_L_Pure_Alcohol', 'Lifeexpectancy':'Life_Expectancy (LED)', 'AdultMortality':'Adult_Mortality', 'Incomecompositionofresources':'HDI',
       'Regional indicator':'Region', 'Ladder score':'Happiness_Score', 'Logged GDP per capita':'GDP',
       'Social support':'Social_Support', 'Healthy life expectancy':'Life_Expectancy (Happiness)',
       'Freedom to make life choices':'Freedom_Score',
       'Perceptions of corruption':'Corruption_Score', 'Ladder score in Dystopia':'Happiness (Dystopia)',
    'percentageexpenditure':'GDP_Health_Gov_Expenditure (%)'
    
})
combined_data
```

```python
combined_data.to_csv('Resources/Data/combined_data.csv')
```
