```python
 %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
 from sklearn.datasets import make_regression
```

```python
 # Read the csv file into a pandas DataFrame

combined = pd.read_csv('Resources/Data/combined_data.csv')
combined.head()

combined = combined.rename(columns={'Unnamed: 0':'country_id'})

combined.head(200)
```

```python
# Steps to replace NaN values:
    
For the whole DataFrame using pandas: df.fillna(0), mean or median 
For the whole DataFrame using numpy: df.replace(np.nan, 0)
    
```

```python
combined.columns
```

```python
# Assign the data to X and y

X = combined[['country_id', 'beer_servings', 'spirit_servings', 'wine_servings',
       'total_litres_of_pure_alcohol', 'Lifeexpectancy',
       'AdultMortality', 'infantdeaths', 'Alcohol', 'percentageexpenditure',
       'HepatitisB', 'Measles', 'BMI', 'under-fivedeaths', 'Polio',
       'Totalexpenditure', 'Diphtheria', 'HIV/AIDS', 'GDP', 'Population',
       'thinness1-19years', 'thinness5-9years', 'Incomecompositionofresources',
       'Schooling',
       'Logged GDP per capita', 'Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption', 'Ladder score in Dystopia']]
y = combined["Ladder score"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
# Use train_test_split to create training and testing data

### BEGIN SOLUTION
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

### END SOLUTION
```

```python
# Create the model using LinearRegression

### BEGIN SOLUTION
from sklearn.linear_model import LinearRegression
model = LinearRegression()
### END SOLUTION
```

```python
# Fit the model to the training data and calculate the scores for the training and testing data

### BEGIN SOLUTION
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```
